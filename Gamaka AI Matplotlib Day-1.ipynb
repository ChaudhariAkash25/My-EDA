```python
import matplotlib.pyplot as plt   #bisic visualization library in python
```

```python
x=[1,2,3]    #x axis values
y=[2,4,6]    #y axis values
plt.plot(x,y,color='b')   #plotting the points
plt.xlabel('x-axis')  #naming the x-axis
plt.ylabel('y-axis')  #naming the y-axis
plt.title('My first graph') #giving a title to my graph
plt.show()   #function to show the graph
```

```python
import numpy as np
N=10
data=plt.np.linspace(0,N,N)
plt.plot(data)
plt.xticks(range(N))  #add loads of ticks
plt.yticks(range(N))  #add loads of ticks
plt.grid()
plt.show()
```

```python
import numpy as np
x=np.array([1,2,3,4])   #x-axis points
y=x*2  #y-axis points
#first plot with x and y axis
plt.plot(x,y)  #plot a chart
x1=[2,4,6,8]
y1=[3,5,7,9]
plt.plot(x1,y1,'-',color='m')    # -(solid line) , -.(dassh-dot line) , :(dotted line)
plt.xlabel('x-axis data')
plt.ylabel('y-axis data')
plt.title('multiple plot')
plt.show()
```

```python
import numpy as np
x=np.array([1,2,3,4])   #x-axis points
y=x*2  #y-axis points
#first plot with x and y axis
plt.plot(x,y)  #plot a chart
x1=[2,4,6,8]
y1=[3,5,7,9]
plt.plot(x1,y1,'-.',)    # -(solid line) , -.(dassh-dot line) , :(dotted line)
plt.xlabel('x-axis data')
plt.ylabel('y-axis data')
plt.title('multiple plot')
plt.fill_between(x, y,y1,color='c',alpha=1)   #alpha blending value,betweeen 0(transperant) and 01(opaque)
plt.show()
```
